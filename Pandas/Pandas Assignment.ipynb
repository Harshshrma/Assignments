4. How many of the columns are numeric and how many of the columns are categorical


```python
numeric,categorical=[],[]
numericc=0
categoricalc=0
for i in df.columns:
    if df[i].dtype in ['int64', 'float64']:
        numericc+=1
        numeric.append(i)
    else:
        categoricalc+=1
        categorical.append(i)
print("Numeric count is ",numericc)
print("Numeric columns are ",numeric,"\n")

print("Categorical count is ",categoricalc)
print("Numeric columns are ",categorical)
```
